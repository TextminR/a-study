### Extracting topics from the 20 newsgroups dataset

```python
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups
```

```python
docs = fetch_20newsgroups(subset='all',  remove=('headers', 'footers', 'quotes'))['data']
```

```python
topic_model = BERTopic()
topics, probs = topic_model.fit_transform(docs)
```

```python
result = topic_model.get_document_info(docs)[["Topic", "Top_n_words", "Probability", "Representative_document"]]
result
```

```python
result.to_csv("topics_20newsgroups.csv")
```
